## Tensor Basics



### Initialization of tensor of different dimensions

```python
import torch

# torch.empty(size): uninitialized
x = torch.empty(1) # scalar
print("Scalar")
print("empty(1):", x)

x = torch.empty(3) # vector, 1D
print("Vector 1D")
print("empty(3):", x)

x = torch.empty(2,3) # matrix, 2D
print("Matrix 2D")
print("empty(2,3):", x)

x = torch.empty(2,2,3) # tensor, 3 dimensions
print("Tensor 3D")
print("empty(2,2,3):", x)

x = torch.empty(2,2,2,3) # tensor, 4 dimensions
print("Tensor 4D")
print("empty(2,2,2,3):", x)
```

### Constructing tensor using values

```python
# torch.rand(size): random number [0,1]
x = torch.rand(5,3)
print("Matrix (5x3) with random number drawn from uniform distribution")
print("rand(5,3):", x)

# torch.zeros(size): fills tensor with zero
# torch.ones(size): fills tensor with one
x = torch.zeros(5,3)
print("Matrix (5x3) with zeros")
print("zeros(5,3):", x)

x = torch.ones(5,3)
print("Matrix (5x3) with ones")
print("ones(5,3):", x)

# check size
print("Size of x:", x.size()) # x.size(0)
print("Size of x:", x.shape) # x.shape[0]

# check data type
print("Data type of x:", x.dtype)

# Build tensor with float16 datatype
x = torch.zeros(5,3, dtype=torch.float16)
print("Matrix (5x3) with zeros with float16")
print("zeros(5,3):", x)
print("Data type of x:", x.dtype)

# Create a tensor with some specific value
x = torch.tensor([5.5, 3])
print("Tensor with specific value")
print("tensor([5.5, 3]):", x)

# Let pytorch know that it needs to track gradient for the given tensor
x = torch.tensor([5.5, 3], requires_grad=True)
print("Tensor with gradient")
print("tensor([5.5, 3], requires_grad=True):", x)
```

### Autograd

This is a cool way in which pytorch collects derivatives through out the computation graph using vector-Jacobian product ([quick refresher](https://www.youtube.com/watch?v=AdV5w8CY3pw)) to calculate final gradient that each parameter will get impacted by. Check out Video 1 of AK () for more on this.

```python
import torch

x = torch.randn(3, requires_grad=True)
y = x+2
print(x) # User created so grad_fn will be none
print(y)
print(y.grad_fn)
z = y*y*2
print(z)
z = z.mean()
print(z)
print(x.grad) # No gradient accumulated until backward is called
z.backward() # dz/dx
print(x.grad) # Since grad class member accumulates gradient for a specific parameter, we need to clear it with zero_grad() after updation and next optimization step
```

```python
# Stop tracking gradient
a = torch.randn(2,2)
b = (a*a).sum()
print(a.requires_grad)
print(b.grad_fn)

a.requires_grad_(True)
b = (a*a).sum()
print(a.requires_grad)
print(b.grad_fn)

#.detach() helps to remove grad for particular values
a = torch.randn(2,2, requires_grad=True)
print(a.requires_grad)
b = a.detach()
print(b.requires_grad)

# Wrap no grad to a set of computation
a = torch.randn(2,2, requires_grad=True)
print(a.requires_grad)
with torch.no_grad():
  b = (a*a).sum()
  print(b.requires_grad)
```

### Gradient Descent Autograd

$f(x) = w * x + b$

where $f(x) = 2 * x$

```python
import torch

X = torch.tensor([1,2,3,4], dtype=torch.float32) # Creating train set
Y = torch.tensor([2,4,6,8], dtype=torch.float32) # Creating test set

w = torch.tensor(0.0, dtype=torch.float32, requires_grad=True) # Initialize weight, initially zero

# Forward pass
def forward(x):
  return w * x

# Calculate Loss
def loss(y, y_pred):
  return ((y_pred - y)**2).mean()

X_test = 5.0
print(f'Prediction before training: f({X_test}) = {forward(X_test).item():.3f}')

# Training
lf = 0.01
n_epochs = 100
for epoch in range(n_epochs):
  # Forward Pass
  y_pred = forward(X) # 1d -> 1d

  # Loss
  l = loss(Y, y_pred) # 1d -> 1 scalar

  # Backward Gradient Calculation
  l.backward() # dl/dw

  # Update weights
  with torch.no_grad():
    w -= lf * w.grad

  # Zero out all the accumulated gradient before next epoch
  w.grad.zero_()

  if epoch % 10 == 0: # Print loss every 10th iteration
    print(f'epoch {epoch+1}: w = {w.item():.3f}, loss = {l.item():.8f}')

print(f'Prediction after training: f({X_test}) = {forward(X_test).item():.3f}')
```

### Typical Pytorch Pipeline

*   Design the model architecture (input, hidden, output layers)
*   Construct Loss and Optimizer
*   Train Loop
    * Forward Pass = Compute Prediction and Loss
    * Backward Pass = Compute Gradients
    * Update Weights using learning rate

```python
import torch
import torch.nn as nn

X = torch.tensor([[1], [2], [3], [4]], dtype=torch.float32) # Creating train set
Y = torch.tensor([[2], [4], [6], [8]], dtype=torch.float32) # Creating test set

n_samples, n_features = X.shape
print(f'n_samples = {n_samples}, n_features = {n_features}')

X_test = torch.tensor([5], dtype=torch.float32) # Test Sample
```

```python
# Step 1 - Designing the model architecture

class LinearRegression(nn.Module): # Simple linear model with one input and one output layer
  def __init__(self, input_dim, output_dim):
    super(LinearRegression, self).__init__()
    # define layers
    self.lin = nn.Linear(input_dim, output_dim)

  def forward(self, x):
    return self.lin(x)

input_size = n_features
output_size = n_features
model = LinearRegression(input_size, output_size)
print(f'Prediction before training: f({X_test.item()}) = {model(X_test).item():.3f}')

# Step 2 - Constructing loss and optimizer
lr = 0.01
n_epochs = 100
loss = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=lr)

# Step 3 - Training Loop
for epoch in range(n_epochs):
  y_pred = model(X) # (n_sample,n_feature) -> (n_sample)
  l = loss(Y, y_pred) # (n_sample) -> scalar
  l.backward() # calculate gradient
  optimizer.step() # Update weight
  optimizer.zero_grad() # Zero out all the accumulated gradient before next epoch

  if (epoch+1) % 10 == 0:
    w, b = model.parameters()
    print(f'epoch {epoch+1}: w = {w[0][0].item():.3f}, loss = {l:.8f}')
print(f'Prediction after training: f({X_test.item()}) = {model(X_test).item():.3f}')
```

### Neural Net Example for MNIST

```python
import torch
import torch.nn as nn
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt

# Use gpu device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Hyperparameters
input_size = 784 # 28x28
hidden_size = 100
num_classes = 10
num_epochs = 2
batch_size = 100
learning_rate = 0.001

# Get MNIST dataset
train_dataset = torchvision.datasets.MNIST(root='./data', train=True, transform=transforms.ToTensor(), download=True)
test_dataset = torchvision.datasets.MNIST(root='./data', train=False, transform=transforms.ToTensor())

# Create dataloaders, these help us get data for each epoch of correct batch size and shuflfes the,
train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)
test_loader = torch.utils.data.DataLoader(dataset=test_dataset, batch_size=batch_size, shuffle=False)

# Lets lookat some data in test dataset
examples = iter(test_loader)
example_data, example_targets = next(examples)

for i in range(6):
  plt.subplot(2,3,i+1) # This just so we can show our 6 examples in a 2x3 grid and mention which example number it is
  plt.imshow(example_data[i][0], cmap='gray')
plt.show()
```

```python
# Step 1
class NeuralNet(nn.Module):
  def __init__(self, input_size, hidden_size, num_classes): # input -> hidden -> output
    super(NeuralNet, self).__init__()
    self.l1 = nn.Linear(input_size, hidden_size)
    self.relu = nn.ReLU()
    self.l2 = nn.Linear(hidden_size, num_classes)

  def forward(self, x):
    out = self.l1(x)
    out = self.relu(out)
    out = self.l2(out)
    return out

model = NeuralNet(input_size, hidden_size, num_classes).to(device)
# input - (batch, 784)
# hidden - (batch, 784) * (784, 100) + (100) -> (batch, 100)
# output - (batch, 100) * (100, 10) + (10) -> (batch, 10)

# Step 2
criterion = nn.CrossEntropyLoss() # categorical cross entropy is cross entropy, basically softmax over logits and then negative log likelihood
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate) # Better than SGD since Adam adapts lr based on tracking prev loss

# Step 3
n_total_steps = len(train_loader)
for epoch in range(num_epochs):
  for i, (images, labels) in enumerate(train_loader):
    # reshape images
    images = images.reshape(-1, 28*28).to(device) # Squashing 2d image to 1d
    labels = labels.to(device)

    # forward pass
    outputs = model(images)
    loss = criterion(outputs, labels)

    # backward pass
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (i+1) % 100 == 0:
      print(f'epoch {epoch+1}/{num_epochs}, step {i+1}/{n_total_steps}, loss = {loss.item():.4f}')
```

```python
# Get test accuracy
with torch.no_grad():
  n_correct = 0
  n_samples = len(test_loader.dataset)
  for images, labels in test_loader:
    images = images.reshape(-1, 28*28).to(device)
    labels = labels.to(device)
    outputs = model(images)
    _, predictions = torch.max(outputs, 1)
    n_correct += (predictions == labels).sum().item()
  acc = 100.0 * (n_correct / n_samples)
  print(f'Accuracy = {acc}')
```

### Convolutional Neural Newtork on CIFAR10

Consists of stacking concolutional and maxpool layers
* Convolutional Layer - Convolutional layer apply kernel/filter of nxn size which slides over the image to take dot product to basically get features
* Max Pooling Layer - Down samples the image by takig max value in a nxn patch

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
import torchvision.transforms as transforms
from torchsummary import summary
import matplotlib.pyplot as plt
import numpy as np

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Hyper parameters
num_epochs = 10
batch_size = 32
learning_rate = 0.001

# Transform base image of type PILImage (pixel value b/w 0 to 1) to normalized image tensor (pixel value b/w -1 to 1), normalization reduces skewness in data
tranform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5, 0.5,0.5), (0.5,0.5,0.5))]) # (image-mean)/std, mean = 0.5 and std = 0.5 for all 3 channels rgb

# Get CIFAR10 dataset and create dataloaders
train_dataset = torchvision.datasets.CIFAR10(root='./data', train=True, transform=tranform, download=True)
test_dataset = torchvision.datasets.CIFAR10(root='./data', train=False, transform=tranform)

train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)
test_loader = torch.utils.data.DataLoader(dataset=test_dataset, batch_size=batch_size, shuffle=False)

classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')

# Lets look at some data
def imshow(img):
  img = img / 2 + 0.5 # unnormalize
  npimg = img.numpy()
  plt.imshow(np.transpose(npimg, (1,2,0)))
  plt.show()

dataiter = iter(train_loader)
images, labels = next(dataiter)
img_grid = torchvision.utils.make_grid(images[0:25], nrow=5) # nrow is number of images per row
imshow(torchvision.utils.make_grid(images))
```

```python
# Step 1
class ConvNet(nn.Module):
  def __init__(self):
    super().__init__()
    self.conv1 = nn.Conv2d(3, 32, 3) # 3 input channels, 32 output channels (32 different filter), 3 filter size (3x3), param=32*(3*3*3+1)=896
    self.pool = nn.MaxPool2d(2,2) # 2x2 area take the max value
    self.conv2 = nn.Conv2d(32, 64, 3) # 32 input channels, 64 output channels, 3 filter, param=64*(3*3*32+1)=18496
    self.conv3 = nn.Conv2d(64, 64, 3) # 64 input channels, 64 output channels, 3 filter, param=64*(3*3*64+1)=36928
    self.fc1 = nn.Linear(64*4*4, 64) # 64*4*4 input, 64 hidden, param=64*4*4*64+64=65600
    self.fc2 = nn.Linear(64, 10) # 64 hidden, 10 output classes, param=64*10+10=650

  def forward(self, x):
    # N, 3, 32, 32 (batch, rgb channel, length px, width px)
    x = F.relu(self.conv1(x))         # (N, 32, 30, 30) (batch, out chan/no of filter, org size - filtersize + 1, org size - filtersize + 1) last two are image dim
    x = self.pool(x)                  # (N, 32, 15, 15) (batch, out chan/no of filter, len / pool len, wid / pool wid)
    x = F.relu(self.conv2(x))         # (N, 64, 13, 13)
    x = self.pool(x)                  # (N, 64, 6, 6)
    x = F.relu(self.conv3(x))         # (N, 64, 4, 4)
    x = torch.flatten(x, start_dim=1) # (N, 1024)
    x = F.relu(self.fc1(x))           # (N, 64)
    x = self.fc2(x)                   # (N, 10)
    return x

model = ConvNet().to(device)
print(summary(model, (3,32,32)))

# Step 2
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

# Step 3
n_total_steps = len(train_loader)
for epoch in range(num_epochs):
  running_loss = 0.0
  for i, (images, labels) in enumerate(train_loader):
    images = images.to(device)
    labels = labels.to(device)

    # Forward Pass
    outputs = model(images)
    loss = criterion(outputs, labels)

    # Backward Pass
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    running_loss += loss.item()
  print(f'epoch {epoch+1}/{num_epochs}, loss = {running_loss/n_total_steps:.4f}')
PATH = './cnn.pth'
torch.save(model.state_dict(), PATH)
```

```python
loaded_model = ConvNet()
loaded_model.load_state_dict(torch.load(PATH, weights_only=True))
loaded_model.to(device)
loaded_model.eval()

# Get testing loss
with torch.no_grad():
  n_correct = 0
  n_correct2 = 0
  n_samples = len(test_loader.dataset)
  for images, labels in test_loader:
    images = images.to(device)
    labels = labels.to(device)
    outputs = model(images)
    _, predictions = torch.max(outputs, 1)
    n_correct += (predictions == labels).sum().item()
    outputs2 = loaded_model(images)
    _, predictions2 = torch.max(outputs2, 1)
    n_correct2 += (predictions2 == labels).sum().item()
  acc = 100.0 * (n_correct / n_samples)
  acc2 = 100.0 * (n_correct2 / n_samples)
print(f'Accuracy = {acc}')
print(f'Accuracy of loaded model = {acc2}')
```
